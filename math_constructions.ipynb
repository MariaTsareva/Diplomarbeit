# PMI

```python
import nltk
import math
def pmi(word1, word2, unigram_freq, bigram_freq, unigram_freq_values, bigram_freq_values):
    prob_word1 = unigram_freq[word1] / float(sum(unigram_freq_values))
    prob_word2 = unigram_freq[word2] / float(sum(unigram_freq_values))
    prob_word1_word2 = bigram_freq / float(sum(bigram_freq_values))
    pmi =  math.log(prob_word1_word2/float(prob_word1*prob_word2),2)
#     print(pmi)
    return pmi

unigrams = nltk.FreqDist(a1)
bigrams = ngrams(a1,2)

n1_freq = nltk.FreqDist(unigrams)
n2_freq = nltk.FreqDist(bigrams)
pmi2 = []
# output_pmi = "C://Users/Maria Tsareva/Desktop/diplom/corp_lem_out.txt"
for bigram, freq in n2_freq.most_common(1000):
    w1 = bigram[0]
    w2 = bigram[1]
    unigram_freq_val = n1_freq.values()
    bigram_freq_val = n2_freq.values()
    pmi_ = pmi(w1, w2, unigrams, freq, unigram_freq_val, bigram_freq_val)
    pmi2.append(pmi_)
#     print(bigram, pmi_)
```

```python
def pmi_tri(word1, word2, word3, unigram_freq, trigram_freq, unigram_freq_values, trigram_freq_values):
    prob_word1 = unigram_freq[word1] / float(sum(unigram_freq_values))
    prob_word2 = unigram_freq[word2] / float(sum(unigram_freq_values))
    prob_word3 = unigram_freq[word3] / float(sum(unigram_freq_values))
    prob_word1_word2_word3 = trigram_freq / float(sum(trigram_freq_values))
    pmi =  math.log(prob_word1_word2_word3/float(prob_word1*prob_word2*prob_word3),2)
#     print(pmi)
    return pmi

unigrams = nltk.FreqDist(a1)
trigrams = ngrams(a1,3)
trigrams_ = []
freqs_3 = []
n1_freq = nltk.FreqDist(unigrams)
n3_freq = nltk.FreqDist(trigrams)
pmi3 = []
# output_pmi = "C://Users/Maria Tsareva/Desktop/diplom/corp_lem_out.txt"
for trigram, freq in n3_freq.most_common(1000):
    trigrams_.append(' '.join(trigram))
    freqs_3.append(freq)
    w1 = trigram[0]
    w2 = trigram[1]
    w3 = trigram[2]
    unigram_freq_val = n1_freq.values()
    trigram_freq_val = n3_freq.values()
    pmi_ = pmi_tri(w1, w2, w3, unigrams, freq, unigram_freq_val, trigram_freq_val)
    pmi3.append(pmi_)
```

```python
def pmi_quad(word1, word2, word3, word4, unigram_freq, quadgram_freq, unigram_freq_values, quadgram_freq_values):
    prob_word1 = unigram_freq[word1] / float(sum(unigram_freq_values))
    prob_word2 = unigram_freq[word2] / float(sum(unigram_freq_values))
    prob_word3 = unigram_freq[word3] / float(sum(unigram_freq_values))
    prob_word4 = unigram_freq[word4] / float(sum(unigram_freq_values))
    prob_4 = quadgram_freq / float(sum(quadgram_freq_values))
    pmi =  math.log(prob_4/float(prob_word1*prob_word2*prob_word3*prob_word4),2)
#     print(pmi)
    return pmi

unigrams = nltk.FreqDist(a1)
quadgrams = ngrams(a1,4)

n1_freq = nltk.FreqDist(unigrams)
n4_freq = nltk.FreqDist(quadgrams)
pmi4 = []
# output_pmi = "C://Users/Maria Tsareva/Desktop/diplom/corp_lem_out.txt"
for quadgram, freq in n4_freq.most_common(1000):
    w1 = quadgram[0]
    w2 = quadgram[1]
    w3 = quadgram[2]
    w4 = quadgram[3]
    unigram_freq_val = n1_freq.values()
    quadgram_freq_val = n4_freq.values()
    pmi_ = pmi_quad(w1, w2, w3, w4, unigrams, freq, unigram_freq_val, quadgram_freq_val)
    pmi4.append(pmi_)
```

```python
def pmi_5(word1, word2, word3, word4, word5, unigram_freq, fivegram_freq, unigram_freq_values, fivegram_freq_values):
    prob_word1 = unigram_freq[word1] / float(sum(unigram_freq_values))
    prob_word2 = unigram_freq[word2] / float(sum(unigram_freq_values))
    prob_word3 = unigram_freq[word3] / float(sum(unigram_freq_values))
    prob_word4 = unigram_freq[word4] / float(sum(unigram_freq_values))
    prob_word5 = unigram_freq[word5] / float(sum(unigram_freq_values))
    prob_5 = fivegram_freq / float(sum(fivegram_freq_values))
    pmi =  math.log(prob_5/float(prob_word1*prob_word2*prob_word3*prob_word4*prob_word5),2)
#     print(pmi)
    return pmi

unigrams = nltk.FreqDist(a1)
fivegrams = ngrams(a1,5)

n1_freq = nltk.FreqDist(unigrams)
n5_freq = nltk.FreqDist(fivegrams)
pmi5 = []
# output_pmi = "C://Users/Maria Tsareva/Desktop/diplom/corp_lem_out.txt"
for fivegram, freq in n5_freq.most_common(1000):
    w1 = fivegram[0]
    w2 = fivegram[1]
    w3 = fivegram[2]
    w4 = fivegram[3]
    w5 = fivegram[4]
    unigram_freq_val = n1_freq.values()
    fivegram_freq_val = n5_freq.values()
    pmi_ = pmi_5(w1, w2, w3, w4, w5, unigrams, freq, unigram_freq_val, fivegram_freq_val)
    pmi5.append(pmi_)
```

```python
def chi_sq(len_d, freq, freq_w1_, freq_w2_):
    chi = len_d*(freq*(len_d - freq_w1_ - freq_w2_) - freq_w1_ * freq_w2_)**2/(freq + freq_w1_)* \
    (freq + freq_w2_)*((len_d - freq_w1_ - freq_w2_) + freq_w1_)*((len_d - freq_w1_ - freq_w2_) + freq_w2_)
    return chi
chis3 = []
unigrams = nltk.FreqDist(a1)
bigrams = ngrams(a1,2)
trigrams = ngrams(a1,3)

n1_freq = nltk.FreqDist(unigrams)
n2_freq = nltk.FreqDist(bigrams)
n3_freq = nltk.FreqDist(trigrams)
len_d = len(a1)
for trigram, freq in n3_freq.most_common(1000):
    w1 = trigram[0]
    w2 = trigram[1]
    w3 = trigram[2]
    tr1 = (w1, w2)
    tr2 = (w2, w3)
    freq_tr1 = n2_freq[tr1]
    freq_tr2 = n2_freq[tr2]
    a = freq_tr1 + freq_tr2
    chi = chi_sq(len_d, a, freq_tr1, freq_tr2)
    chis3.append(chi)
    
```

```python
def llr(a, b, c, d):
    result = 2 * (a * math.log(a) + b * math.log(b) + c * math.log(c)
                  + d * math.log(d)
                  - (a + b) * math.log(a + b) - (a + c) * math.log(a + c)
                  - (b + d) * math.log(b + d) - (c + d) * math.log(c + d) 
                  + (a + b + c + d) * math.log(a + b + c + d))
    return result

llrs2 = []
unigrams = nltk.FreqDist(a1)
bigrams = ngrams(a1,2)
bigrams_ = []
freqs_2 = []
n1_freq = nltk.FreqDist(unigrams)
n2_freq = nltk.FreqDist(bigrams)
len_d = len(a1)
print(len_d)
# output_pmi = "C://Users/Maria Tsareva/Desktop/diplom/corp_lem_out.txt"
for bigram, freq in n2_freq.most_common(1000):
    bigrams_.append(' '.join(bigram))
    freqs_2.append(freq)
    w1 = bigram[0]
    w2 = bigram[1]
    freq_w1 = n1_freq[w1]
    freq_w2 = n1_freq[w2]
    freq_w1_ = freq_w1 - freq
    freq_w2_ = freq_w2 - freq
    d = len_d-freq_w1_-freq_w2_
    try:
        llr_ = llr(freq, freq_w1_, freq_w2_, d)
        llrs2.append(llr_)
    except ValueError:
        llr_ = 1000
        llrs2.append(llr_)
```

```python
def count_c_value_for_ngrams(d, df, d_for_ngrams):
    for i in tqdm_notebook(df.index):
        
        if not d.get(df['ngram'].loc[i]):
            df['c-value'].loc[i] = np.log2(2) * df['freq'].loc[i]
        else:
            df['c-value'].loc[i] = np.log2(2) * (df['freq'].loc[i] - d[df['ngram'].loc[i]]['freq'] /
                                                 len(d[df['ngram'].loc[i]]['ngram']))
            d_for_ngrams = count_c_value_for_bigger_ngrams(d_for_ngrams, d, df, i)
    
    return df, d_for_ngrams

def count_c_value_for_bigger_ngrams(d_for_ngrams, d, df, i):
    for elem in d[df['ngram'].loc[i]]['ngram']:
        bigger_ngrams = {'freq': 0, 'ngram': {}}
        
        for b_elem in d[df['ngram'].loc[i]]['ngram']:
            if elem in b_elem and elem != b_elem:
                bigger_ngrams['freq'] += d[df['ngram'].loc[i]]['ngram'][b_elem]
                bigger_ngrams['ngram'][b_elem] = d[df['ngram'].loc[i]]['ngram'][b_elem]
        
        d_for_ngrams['ngram'].append(elem)
        
        if bigger_ngrams['freq'] != 0:
            d_for_ngrams['c-value'].append(np.log2(len(elem.split()) * (d[df['ngram'].loc[i]]['ngram'][elem] -
                                             bigger_ngrams['freq']/len(bigger_ngrams['ngram']))))
        else:
            d_for_ngrams['c-value'].append(np.log2(len(elem.split()) * d[df['ngram'].loc[i]]['ngram'][elem]))
    
    return d_for_ngrams

def create_w_context(c_value_250, text):
    w_context = defaultdict(lambda: defaultdict(lambda: 0))            
    for elem in c_value_250['ngram']:
#         print(elem)
        pat = re.compile('(\w+ )?{} (\w+)?'.format(elem))
        res = re.findall(pat, text)
        for words in res:
            for w in words:
                w_context[elem][w] += 1
    return w_context

def create_w_count(w_context):
    w_count = defaultdict(lambda: 0)
    
    for ngram in w_context:
        words = []
        for w in w_context[ngram]:
            if w not in words:
                w_count[w] += 1
                words.append(w)                
    return w_count

def count_nc_value(i, w_context, w_count, c_value_250):
    context_words = 0    
    for elem in set(w_context[c_value_250['ngram'].loc[i]]):
        count_elem = 0        
        for other in w_context[c_value_250['ngram'].loc[i]]:
            if elem == other:
                count_elem += 1        
        context_words += count_elem * w_count[elem] / len(w_count)    
    return 0.8 * c_value_250['c-value'].loc[i] + 0.2 * context_words

def count_c_nc_value(df_2, d):
    
#     df = pd.read_csv('{}/bigrams_{}.txt'.format(PATH_bigrams, pos), sep='\t',
#                    names=['id', 'freq', 'range', 'ngram'], index_col=False)
    
    df_2['c-value'] = 0
    d_for_ngrams = {'ngram': [], 'c-value': []}    
    df, d_for_ngrams = count_c_value_for_ngrams(d, df_2, d_for_ngrams)    
    c_value_all = pd.concat([df[['ngram','c-value']],
                             pd.DataFrame(d_for_ngrams)]).sort_values('c-value', ascending=False)
    c_value_all.to_csv('cvalue_2_x.txt', sep='\t')
    c_value_250 = c_value_all[0:249]
    w_context = create_w_context(c_value_250, text)
    w_count = create_w_count(w_context)
    c_value_250['nc-value'] = 0
    for i in c_value_250.index:
        c_value_250['nc-value'].loc[i] = count_nc_value(i, w_context, w_count, c_value_250)
    c_value_250.sort_values('nc-value', ascending=False).to_csv('ncvalue_2_x.txt',
                           sep='\t', index=False)

    
count_c_nc_value(df_2, d)
```
